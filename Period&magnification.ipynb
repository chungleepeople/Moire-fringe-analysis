```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import cv2
import os
from PIL import Image
import math
```

```python
p1=4
xs=[]
p_moires=[]

for i in np.arange(1.01,2,0.01):
    p_moire=p1*i/(i-1)
    p_moires.append(p_moire)
    xs.append(i)
# plt.xticks(np.linspace(1, 2, 10))

p1=5
xd=[]
p_moired=[]

for i in np.arange(1.01,2,0.01):
    p_moire=p1*i/(i-1)
    p_moired.append(p_moire)
    xd.append(i)

plt.figure(figsize=(16,9))
plt.plot(xd,p_moired,'r--',linewidth=2,label='$p_\mathrm{1}$=5 (µm)')
plt.plot(xs,p_moires,'b-',linewidth=2,label='$p_\mathrm{1}$=4 (µm)')

plt.xlabel(r'$ \alpha$',fontsize=20)
plt.ylabel('Period of moire fringe (µm) ',fontsize=20)
plt.legend(loc='upper right',fontsize=20)  
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True)
plt.tick_params('both',direction='in')
    


```
